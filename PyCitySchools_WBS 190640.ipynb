```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# DISTRICT SUMMARY


TotSchools = len(school_data["school_name"])
#print('Total Schools: ' + str(TotSchools))

TotStudents = len(student_data["student_name"])
#print('Total Students:  ' + str(TotStudents))

TotBudget = sum(school_data["budget"])
#print('Total Budget:  ' + str(TotBudget))

AvgMath = (sum(student_data["math_score"]) / len(student_data["math_score"]))
#print('Average Math Score:  ' + str(AvgMath))

AvgReading = (sum(student_data["reading_score"]) / len(student_data["reading_score"]))
#print('Average Reading Score:  ' + str(AvgReading))

OverallPassRate = (AvgMath + AvgReading) / 2
#print('Overall Pass Rate:  ' + str(OverallPassRate))

PassingMath = len(student_data[student_data['math_score'] >= 70])
#print(PassingMath)
PassingMathPct = PassingMath/TotStudents*100
#print('Passing Math %: ' + str(PassingMathPct))

PassingReading = len(student_data[student_data['reading_score'] >= 70])
#print(PassingReading)
PassingReadingPct = PassingReading/TotStudents*100
#print('Passing Reading %:  ' + str(PassingReadingPct))

district_summary = [[TotSchools, TotStudents, TotBudget, AvgMath, AvgReading,  \
                   PassingMathPct, PassingReadingPct, OverallPassRate]]
                    

district_summary_df = pd.DataFrame(district_summary, columns = ['Total Schools', \
                    'Total Students', 'Total Budget', 'Average Math Score', \
                    'Average Reading Score', '% Passing Math', '% Passing Reading', \
                    '% Overall Pass Rate'])

district_summary_df['Total Budget']= district_summary_df['Total Budget'].map('${:,.2f}'.format)

district_summary_df


```

```python


PMath = school_data_complete[(school_data_complete["math_score"] >= 70)]
perschool_passing_math = PMath.groupby(["school_name"]).count()["student_name"]/school_data_complete["school_name"].value_counts()*100

#perschool_passing_math

PReading = school_data_complete[(school_data_complete["reading_score"] >= 70)]
perschool_passing_reading = PReading.groupby(["school_name"]).count()["student_name"]/school_data_complete["school_name"].value_counts()*100

#perschool_passing_math


perschool_passing_overall = (perschool_passing_math + perschool_passing_reading) / 2

perschool_passing_overall

#district_summary = [[TotSchools, TotStudents, TotBudget, AvgMath, AvgReading,  \
#                   PassingMathPct, PassingReadingPct, OverallPassRate]]
```
